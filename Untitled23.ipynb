```python
# This function will CLEAN the GDP data
def clean_gdp(df):
    # Make a copy of the original data
    # (so we don't accidentally change gdp_df)
    df = df.copy()

    # 1. Remove extra spaces from the country names
    df['Country/Territory'] = df['Country/Territory'].str.strip()

    # 2. Remove extra spaces from ALL text columns
    df = df.apply(
        lambda col: col.str.strip() if col.dtype == "object" else col
    )

    # 3. Send back (return) the cleaned data
    return df
```
